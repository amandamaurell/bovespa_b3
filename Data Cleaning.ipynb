```python
# 'HistoricoFormularioReferencia.aspx?codigoCVM={code}&tipo=itr&ano=0' 
import pandas as pd
import numpy as np
```

```python
empresas_df = pd.read_csv('url_demonstrativos.csv')
receitas_1_df = pd.read_excel('Receitas_1.xlsx')
receitas_2_df = pd.read_excel('Receitas_2.xlsx')
receitas_3_df = pd.read_excel('Receitas_3.xlsx')
lucros_1_df = pd.read_excel('Lucros_1.xlsx')
lucros_2_df = pd.read_excel('Lucros_2.xlsx')
lucros_3_df = pd.read_excel('Lucros_3.xlsx')
```

```python
receitas_1_df.head()
```

```python
empresas_df = empresas_df.rename(columns={'Unnamed: 0':'Empresas'})
```

```python
columns_to_drop = ['01/01/2020 a 30/06/2020','01/01/2019 a 30/06/2019']
receitas_2_df.drop(columns=columns_to_drop, inplace=True)
lucros_2_df.drop(columns=columns_to_drop, inplace=True)
```

```python
receitas_3_df.head(2)
```

```python
columns_to_drop = ['01/01/2020 a 30/09/2020','01/01/2019 a 30/09/2019']
receitas_3_df.drop(columns=columns_to_drop, inplace=True)
lucros_3_df.drop(columns=columns_to_drop, inplace=True)
```

```python
receitas_df = pd.concat([empresas_df['Empresas'],receitas_1_df, receitas_2_df[['01/04/2020 a 30/06/2020','01/04/2019 a 30/06/2019']], receitas_3_df[['01/07/2020 a 30/09/2020','01/07/2019 a 30/09/2019']]], axis=1).set_index('Empresas')
```

```python
receitas_df.head(2)
```

```python
lucros_df = pd.concat([empresas_df['Empresas'],lucros_1_df, lucros_2_df[['01/04/2020 a 30/06/2020','01/04/2019 a 30/06/2019']], lucros_3_df[['01/07/2020 a 30/09/2020','01/07/2019 a 30/09/2019']]], axis=1).set_index('Empresas')
```
