```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup as bsp
```

```python
pattern = re.compile(r'codigoCvm=\d+')
numbers = {}
with open('bovespa.txt', 'r', encoding='utf-8', errors='ignore') as file:
    f = file.read()
    numbers['code'] = re.findall(pattern, f)
    df= pd.DataFrame(numbers)
df = df.drop_duplicates()
df['code'] = df['code'].map(lambda x: x.strip('codigoCvm='))
```

```python
companies = {}
for i in codigos['code'][359:]:
    url_relatorios = f'http://bvmf.bmfbovespa.com.br/cias-listadas/empresas-listadas/ResumoDemonstrativosFinanceiros.aspx?codigoCvm={i}&idioma=pt-br'
    response = requests.get(url_relatorios)
    soup = bsp(response.content, "html.parser")
    content_one = soup.find('div', id='containerDemonstrativos')
    if content_one == None:
        continue
    else:
        content = content_one.find_all('a')
        name = soup.find('span', id='ctl00_contentPlaceHolderConteudo_MenuEmpresasListadas1_lblNomeEmpresa').string
        companies[name] = []
        for i in content[0:3]:
            companies[name].append(i.get('href').lstrip("javascript:AbreFormularioCadastral('").rstrip("')"))

companies
```

```python
companies_df = pd.DataFrame(companies).T
companies_df.to_csv('url_demonstrativos.csv', mode='a')
```
